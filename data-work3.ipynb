```python
# Import seaborn and apply its plotting styles
import seaborn as sns
sns.set(style="white", font_scale=1.5)

# import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
# set plotting size parameter
plt.rcParams['figure.figsize'] = (17, 7)

# make sure pandas & numpy are imported
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

#improve resolution
#comment this line if erroring on your machine/screen
%config InlineBackend.figure_format ='retina'
```

```python
df = pd.read_csv("data.csv")
```

### Behavioral Patterns Distribution

```python
beh_probs = df["behavior_problems_prevalence"][df["behavior_problems_prevalence"].notnull()]
prop_beh_probs = beh_probs.value_counts()/len(beh_probs)
sns.barplot(x=beh_probs.unique(), y=prop_beh_probs, color="#9B5DD1");
plt.ylabel("Proportion")
plt.xlabel("Behavioral Patterns and Problems")
plt.xticks((0, 1, 2), ("No behavioral problems", "Told, but no longer have behavior problems", "Has behavioral problems"))
plt.title("Behavioral Patterns and Prevalence")
plt.show()
```

### Homework Completion Distribution

```python
does_hw = df["does_hw"][df["does_hw"].notnull()]
prop_does_hw = does_hw.value_counts()/len(does_hw)
sns.barplot(x=does_hw.unique(), y=prop_does_hw, color="#65BDD0");
plt.ylabel("Proportion")
plt.xlabel("Homework Completion")
plt.xticks((0, 1, 2, 3,4), ("Do not have hw (age 0-5)", "Always does hw", "Usually does hw",'Sometimes does hw','Never does hw'))
plt.title("Homework Completion")
plt.show()
```

### Gender Distribution

```python
gen = df["gender"][df["gender"].notnull()]
prop_gen = gen.value_counts()/len(gen)
sns.barplot(x=gen.unique(), y=prop_gen, color="#65D069");
plt.ylabel("Proportion")
plt.xlabel("Gender")
plt.xticks((0, 1), ("Male", "Female"))
plt.title("Gender Distribution")
plt.show()
```

### Behavioral Patterns vs. Homework Completion

```python
x = df['does_hw'][df['does_hw'].notnull()]
x = x.unique()
y1 = df[df['behavior_problems_prevalence']==1]
y1 = y1["does_hw"][y1["does_hw"].notnull()].value_counts()/len(y1["does_hw"][y1["does_hw"].notnull()])
y2 = df[df["behavior_problems_prevalence"]==2]
y2 = y2["does_hw"][y2["does_hw"].notnull()].value_counts()/len(y2["does_hw"][y2["does_hw"].notnull()])
y3 = df[df["behavior_problems_prevalence"]==3]
y3 = y3["does_hw"][y3["does_hw"].notnull()].value_counts()/len(y3["does_hw"][y3["does_hw"].notnull()])
temp = pd.DataFrame({"Does not have behavioral problems": y1, "Told, but no longer have behavioral problems": y2,'Has behavioral problems':y3},index=x)
temp.plot.bar(rot=0, color={"Does not have behavioral problems": "#F9A2FA", 
                           "Told, but no longer have behavioral problems": "#C35EDE",
                           'Has behavioral problems':'#7301A5'})
plt.xlabel("Homework Completion")
plt.ylabel("Proportion")
plt.title("Behavior Pattern vs. Homework Completion")
plt.xticks((0, 1, 2, 3, 4), ("Does not have hw (age 0-5)", "Always does hw", "Usually does hw",'Sometimes does hw','Never does hw'))
plt.show()
```

Individuals that do not have behavioral problems had the highest proportion in the 'always does homework' category and the lowest proportion in the other three categories 'usually does homework','sometimes does homework', and 'never does homework.' Individuals with behavioral problems in the past had the highest proportion in the 'usually does homework; category, and the second highest proportion in the 'always does homework','sometimes does homework', and 'never does homework' categories. Individuals described as having behavioral problems currently had the highest proportion in both 'sometimes does homework' and 'never does homework' categories, but the smallest proportion in the 'always does homework' category and the second largest proportion in the 'usually does homework' category.
